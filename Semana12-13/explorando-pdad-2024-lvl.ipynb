```python
import pandas as pd

#Carregando os dados e pegando 1.000 idades
moradores = pd.read_csv("moradores.csv", sep=";", decimal=",", low_memory=False)
validos = moradores[moradores["idade_calculada"] != 99999].head(1000).copy()
idades_base = validos["idade_calculada"].tolist()

# Versões ordenadas e invertidas
idades_ordenadas = sorted(idades_base)
idades_invertidas = sorted(idades_base, reverse=True)

def bubble_sort_conta(lista):
    lst = lista[:]
    comparacoes = 0
    n = len(lst)
    for i in range(n):
        for j in range(n - i - 1):
            comparacoes += 1
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
    return lst, comparacoes

def selection_sort_conta(lista):
    lst = lista[:]
    comparacoes = 0
    n = len(lst)
    for i in range(n):
        idx_min = i
        for j in range(i + 1, n):
            comparacoes += 1
            if lst[j] < lst[idx_min]:
                idx_min = j
        lst[i], lst[idx_min] = lst[idx_min], lst[i]
    return lst, comparacoes

def insertion_sort_conta(lista):
    lst = lista[:]
    comparacoes = 0
    for i in range(1, len(lst)):
        chave = lst[i]
        j = i - 1
        while j >= 0 and lst[j] > chave:
            comparacoes += 1
            lst[j + 1] = lst[j]
            j -= 1
        comparacoes += 1
        lst[j + 1] = chave
    return lst, comparacoes

cenarios = {
    "Lista desordenada": idades_base,
    "Lista ordenada": idades_ordenadas,
    "Lista invertida": idades_invertidas
}

n = len(idades_base)
print(f"Testando algoritmos com {n} elementos:\n")

for nome_cenario, dados in cenarios.items():
    _, c_bubble = bubble_sort_conta(dados)
    _, c_selection = selection_sort_conta(dados)
    _, c_insertion = insertion_sort_conta(dados)
    
    print(nome_cenario)
    print("-" * 40)
    print(f"  Bubble Sort:    {c_bubble} comparações")
    print(f"  Selection Sort: {c_selection} comparações")
    print(f"  Insertion Sort: {c_insertion} comparações\n")
```

```python
# Bubble e Selection Sort têm desempenho fixo, realizando o mesmo número de comparações independentemente da ordem inicial.
# O Insertion Sort é muito eficiente em listas ordenadas, mas atinge seu pior cenário em listas totalmente invertidas.
# Essa diferença prova que a eficiência real de um código depende da interação entre o algoritmo e a natureza dos dados. 
# Portanto, a organização inicial dos elementos é um fator decisivo para o desempenho final do processamento.
```

```python
import pandas as pd

#Carrega os dados
def carregar_dados():
    mor = pd.read_csv("moradores.csv", sep=";", decimal=",", encoding="latin1",engine="python")
    dom = pd.read_csv("domicilios.csv", sep=";", decimal=",", encoding="latin1",engine="python", on_bad_lines='skip')
    return mor, dom

# Modificamos as funções para aceitar um arquivo 'f' opcional
def relatorio_idades(moradores, f=None):
    validos = moradores[moradores["idade_calculada"] != 99999].head(2000).copy()
    idades = validos["idade_calculada"].tolist()
    print(f"\n  Total com idade declarada: {len(idades)}", file=f)
    print(f"  Média: {sum(idades)/len(idades):.1f} anos", file=f)
    print(f"  Mínima: {min(idades)} | Máxima: {max(idades)}", file=f)

def relatorio_escolaridade(moradores, f=None):
    nomes = {1:"Sem instrução",2:"Fund. incompleto",3:"Fund. completo",
             4:"Médio incompleto",5:"Médio completo",6:"Superior incompleto",
             7:"Superior completo",8:"Pós-graduação"}
    contagem = {}
    for _, linha in moradores.iterrows():
        n = linha["escolaridade"]
        if n in nomes:
            contagem[n] = contagem.get(n, 0) + 1
    print("\n  Escolaridade:", file=f)
    for nivel, total in sorted(contagem.items()):
        print(f"    {nomes[nivel]}: {total}", file=f)

def relatorio_domicilios(domicilios, f=None):
    print(f"\n  Total de domicílios: {len(domicilios)}", file=f)
    media_pessoas = sum(domicilios["A01npessoas"]) / len(domicilios)
    print(f"  Média de moradores por domicílio: {media_pessoas:.1f}", file=f)

# Nova função para salvar tudo
def salvar_relatorio_completo(moradores, domicilios):
    with open("relatorio_completo.txt", "w", encoding="utf-8") as f:
        f.write("Relatório PDAD 2024, dos primeiros 2000\n")
        relatorio_idades(moradores, f=f)
        relatorio_escolaridade(moradores, f=f)
        relatorio_domicilios(domicilios, f=f)
    print("\n  Sucesso! Relatório salvo em 'relatorio_completo.txt'")

moradores, domicilios = carregar_dados()

while True:
    print("\n" + "=" * 40)
    print("  PDAD 2024 — Menu de Relatórios")
    print("=" * 40)
    print("  1. Relatório de idades")
    print("  2. Relatório de escolaridade")
    print("  3. Relatório de domicílios")
    print("  4. Salvar relatório completo em TXT")
    print("  0. Sair")
    opcao = input("\n  Escolha uma opção: ").strip()

    if opcao == "1":
        relatorio_idades(moradores)
    elif opcao == "2":
        relatorio_escolaridade(moradores)
    elif opcao == "3":
        relatorio_domicilios(domicilios)
    elif opcao == "4":
        salvar_relatorio_completo(moradores, domicilios)
    elif opcao == "0":
        print("  Até logo!")
        break
    else:
        print("  Opção inválida.")
```

```python
import sys
import pandas as pd
import time

RA_NOMES = {
    5249: "Arniqueira", 5301: "Brasília", 5303: "Taguatinga",
    5305: "Sobradinho", 5311: "Cruzeiro", 5313: "Ceilândia",
    5314: "Sobradinho II", 5315: "Jardim Botânico", 5319: "Lago Sul",
    5320: "Gama", 5326: "Samambaia", 5328: "Santa Maria",
    5330: "São Sebastião",
}

# Filtra o dataset para manter apenas os dados das RAs listadas no dicionário acima
moradores = moradores[moradores["localidade"].isin(ra_nomes.keys())]

ESCOLARIDADE = {
    1:"Sem instrução", 2:"Fund. incompleto", 3:"Fund. completo",
    4:"Médio incompleto", 5:"Médio completo", 6:"Superior incompleto",
    7:"Superior completo", 8:"Pós-graduação"
}

def bubble_sort(lista):
    n = len(lista)
    lista = lista.copy()
    for i in range(n):
        for j in range(n - i - 1):
            if lista[j]["idade"] > lista[j + 1]["idade"]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista

def insertion_sort_escolaridade(lista):
    # Ordena pela chave 'nivel_num' (1 a 8)
    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1
        while j >= 0 and lista[j]['nivel_num'] > chave['nivel_num']:
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = chave
    return lista

def quick_sort(lista):
    if len(lista) <= 1: return lista
    pivo = lista[len(lista) // 2]["idade"]
    esquerda = [x for x in lista if x["idade"] < pivo]
    meio = [x for x in lista if x["idade"] == pivo]
    direita = [x for x in lista if x["idade"] > pivo]
    return quick_sort(esquerda) + meio + quick_sort(direita)

def gerar_conteudo_relatorio(ra_codigo, moradores_df):
    filtro = moradores_df[moradores_df["localidade"] == ra_codigo]
    if filtro.empty: return None

    ra_nome = RA_NOMES.get(ra_codigo, f"RA-{ra_codigo}")
    validos = filtro[filtro["idade_calculada"] != 99999].head(2000).copy()
    
    # Processa Moradores
    lista_moradores = []
    for _, linha in validos.iterrows():
        lista_moradores.append({"id": linha["morador_id"], "idade": linha["idade_calculada"]})
    
    # Ordena por idade 
    lista_ordenada = quick_sort(lista_moradores)

    # Processa Escolaridade 
    contagem = {}
    for n in validos["escolaridade"]:
        if n in ESCOLARIDADE:
            contagem[n] = contagem.get(n, 0) + 1
    
    lista_esc = [{"nivel": ESCOLARIDADE[k], "nivel_num": k, "total": v} for k, v in contagem.items()]
    lista_esc = insertion_sort_escolaridade(lista_esc)

    # Monta o relatório
    linhas = [f"Relatório da RA: {ra_nome}", "="*30]
    linhas.append(f"Total moradores: {len(validos)}")
    
    linhas.append("\nEscolaridade (Ordenada por nível):")
    for item in lista_esc:
        linhas.append(f" - {item['nivel']}: {item['total']}")
        
    return "\n".join(linhas)

#olhando a velocidade
def testar_performance(lista):
    print("\n--- Comparação de Algoritmos (Ordenando por idade) ---")
    
    start = time.time()
    bubble_sort(lista)
    print(f"Bubble Sort: {time.time() - start:.4f}s")
    
    start = time.time()
    quick_sort(lista)
    print(f"Quick Sort: {time.time() - start:.4f}s")

moradores = pd.read_csv("moradores.csv", sep=";", decimal=",", encoding="utf-8-sig", low_memory=False)

# Detecta argumentos
if "--todas" in sys.argv:
    print("Gerando relatório para todas as RAs...")
    for ra_code in RA_NOMES.keys():
        conteudo = gerar_conteudo_relatorio(ra_code, moradores)
        if conteudo:
            nome_arq = f"relatorio_{ra_code}.txt"
            with open(nome_arq, "w", encoding="utf-8") as f:
                f.write(conteudo)
            print(f"Salvo: {nome_arq}")
    
    # Benchmark rápido
    testar_performance(moradores[moradores["idade_calculada"] < 100].head(500).to_dict('records'))

elif len(sys.argv) >= 2:
    ra = int(sys.argv[1])
    print(gerar_conteudo_relatorio(ra, moradores))
else:
    print("Uso: python analise_ra.py <codigo_ra> ou --todas")
```
